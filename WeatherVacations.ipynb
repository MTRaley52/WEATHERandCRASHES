```python
 # Dependencies
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
from config import api_key
```

```python
 # Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#Top Travel Destinations in the United States
cities = ["Houston", "Chicago", "Austin", "Boulder", "New Orleans", "Seattle", 
          "New York", "Nashville", "Savannah", "Austin", "Sacramento", "Boise", "Portland", 
          "Boston", "Aspen", "Park City", "Denver", "Asheville", "Santa Fe", "Madison", 
          "Detroit", "Philadelphia", "Las Vegas", "Nantucket", "Sedona", "Kansas City", 
          "Indianapolis", "Minneapolis", "Manchester", "Miami"]

# Lists to hold Latitudes and Temps
lat = []
temp = []

# Loop through the list of cities and perform a request for data on each
for city in cities:
    response = requests.get(query_url + city).json()
    lat.append(response['coord']['lat'])
    temp.append(response['main']['temp'])

print(f"The latitude information received is: {lat}")
print(f"The temperature information received is: {temp}")
print(len(lat))
print(len(temp))
```

```python
#data frame for City, Latitude, Temperature
weather_dict = {
    "City": cities,
    "Latitude": lat,
    "Temperature": temp
}
weather_data = pd.DataFrame(weather_dict)
weather_data
```
