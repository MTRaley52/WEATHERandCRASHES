```python
lat_lngs = zip(df["LATITUDE"],df["LONGITUD"])
cities = []




# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    cities.append(city)
   
     

```

```python
df["city"] = cities
```

```python
import openweathermapy.core as owm
settings = {
    "units": "metric",
    "appid": api_key
}
```

```python
df2 = df.head(100)
```

```python
cities = df2["city"]
temp = []
for city in cities:
    current_weather = owm.get_current(city, **settings)
    try:
        if current_weather['cod'] != 200:
            temp.append("na")
        else:
            temp.append(current_weather['main']['temp'])
    except HTTPError:
        temp.append("na")
    continue
        
```

```python
df2["temperature"] = temp
df2
```
